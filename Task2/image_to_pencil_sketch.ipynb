# **Importing packages**

```python
import cv2
from google.colab.patches import cv2_imshow
```

# **Loading image**

```python
image = cv2.imread("cat.jpg")
cv2_imshow(image)
cv2.waitKey(0)
```

Converting above image into grayscale image.


```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
cv2.waitKey(0)
```

Reversing the color on image to create negative image ,that is inverted image.

```python
inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
cv2.waitKey()
```

To make blurry image of inverted image we use GaussianBlur() method.

```python
blurred = cv2.GaussianBlur(inverted_image, (21,21), 0)
cv2_imshow(blurred)
```

Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image.

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2_imshow( pencil_sketch)
cv2.waitKey(0)
```

```python
cv2_imshow( image)
cv2_imshow( pencil_sketch)
cv2.waitKey(0)
```
